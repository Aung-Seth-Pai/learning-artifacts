#### Decorator Pattern

The Inheritance Trap
When we need to extend a class's functionality, our instinct is often to use Inheritance. However, this leads to several architectural "code smells":
- **Static Nature**: Inheritance is fixed at compile-time. You cannot "un-inherit" or swap behaviors while the program is running.
- **Class Explosion**: If you have a `Coffee` class and want to add `Milk`, `Sugar`, and `Vanilla`, you end up creating `CoffeeWithMilk`, `CoffeeWithSugar`, `CoffeeWithMilkAndSugar`, etc. Likewise, if we create a `FileStream` class, and then want to add encryption, we need to create `EncryptedFileStream`. To add compression, we create `CompressedFileStream`. If we want both, we are forced to create `EncryptedAndCompressedFileStream`. This scales exponentially and is a maintenance nightmare.
- **Fragile Base Class**: Changes in the parent class ripple down and risk breaking all subclasses.
- **Single Inheritance Limit**: In many languages (like Java or C#), a class can only have one parent.


**Solution: Composition & Delegation**
Instead of saying an object is something (Inheritance), we say it has something (Composition).

Through Delegation, an object receives a request and passes it along to another object that actually knows how to handle it. This allows us to combine multiple behaviors at runtime.

Comparison between Aggregation & Composition
| Feature | Aggregation | Composition |
| :--- | :--- | :--- |
| **Relationship** | **"Weak"** (A uses B) | **"Strong"** (A owns B) |
| **Lifecycle** | The child can exist independently of the parent. | The child is destroyed if the parent is destroyed. |
| **Example** | A Department and Professors. | A House and Rooms. |

#### The Decorator Pattern
The Decorator (also known as a Wrapper) is a structural pattern that lets you attach new behaviors to objects by placing these objects inside special wrapper objects that contain the behaviors.

**How it Works**
- **The Interface**: The wrapper implements the exact same interface as the object it's wrapping.

- **The Delegation**: When the client calls a method on the wrapper, the wrapper performs its "extra" task (e.g., logging, encrypting) and then delegates the call to the wrapped object.

- **Transparency**: Because the interface is identical, the client doesn't know (or care) if they are talking to the original object or a chain of ten decorators.\

**The Components**
To implement this correctly, we follow this structure:
- **Component (Interface)**: The common interface for both the real object and the decorators.
- **Concrete Component**: The basic object that defines the core behavior (e.g., the plain Coffee).
- **Base Decorator**: A class that holds a reference to a Component and implements the interface by delegating to it.
- **Concrete Decorators**: Classes that override methods of the base decorator to add "extra" behavior (e.g., MilkDecorator, SugarDecorator).

```python
# Example

from abc import ABC, abstractmethod

# 1. The Interface
class Coffee(ABC):
    @abstractmethod
    def get_cost(self): pass

# 2. Concrete Component
class SimpleCoffee(Coffee):
    def get_cost(self):
        return 10  # Base price

# 3. Base Decorator
class CoffeeDecorator(Coffee):
    def __init__(self, coffee: Coffee):
        self._decorated_coffee = coffee

    def get_cost(self):
        return self._decorated_coffee.get_cost()

# 4. Concrete Decorators
class Milk(CoffeeDecorator):
    def get_cost(self):
        return super().get_cost() + 2

class Sugar(CoffeeDecorator):
    def get_cost(self):
        return super().get_cost() + 1

# --- Usage (The "Why" in action) ---
# We can mix and match behaviors at runtime without new classes!
my_coffee = SimpleCoffee()
my_coffee = Milk(my_coffee)
my_coffee = Sugar(my_coffee)

print(f"Total Cost: ${my_coffee.get_cost()}") # Total Cost: $13
```

**Why this is better**:
- **Flexibility**: We can add or remove decorators at runtime.

- **Single Responsibility Principle**: You can divide a monolithic class that implements every possible variant into several smaller, focused classes.

- **Recursive Nesting**: You can wrap a decorator inside another decorator infinitely.
