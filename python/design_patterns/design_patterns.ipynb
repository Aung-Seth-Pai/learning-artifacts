## 3 Main Types of Design Patterns

Design patterns are proven solutions to common software design problems. They help improve code reusability, scalability, and maintainability. There are **three main categories** : Creational, Structural and Behavioral.

#### 🧩 When (and When Not) to Use Design Patterns

Design patterns are not rules — they are flexible, reusable solutions to common problems in software design. But using them without context can lead to overengineering. Here's how to evaluate when to apply them.

---

### ✅ **When to Use Design Patterns**

1. **Recurring Design Problems**
   - You've encountered a similar problem before and want a clean, proven solution.

2. **Improving Code Quality**
   - You need better modularity, flexibility, or testability.

3. **Designing for Change**
   - Your system is expected to evolve — patterns help with **modifiability**, **extensibility**, and **maintainability**.

4. **Collaborative Projects**
   - Patterns give developers a shared vocabulary for communicating solutions.

---

### 🚫 **When *Not* to Use Design Patterns**

1. **Premature Optimization**
   - Don't use patterns just because they exist — only when there's a real need.

2. **Simple Problems**
   - Patterns may introduce unnecessary complexity for small, straightforward use cases.

3. **Lack of Clear Requirements**
   - If you don’t understand the problem fully, applying a pattern can create more confusion.

---

### 🧠 How to Decide

Design patterns work best when **forces** and **quality attributes** align with the pattern’s intent:

- **Participants**  
  Identify the classes and objects involved in the problem. Patterns define clear roles and responsibilities.

- **Quality Attributes**  
  Patterns support non-functional goals like **usability**, **performance**, **modifiability**, and **reliability**. For example:
  - Use **Observer** for real-time updates (modifiability)
  - Use **Decorator** to add behavior without modifying code (extensibility)

- **Forces**  
  Patterns help resolve trade-offs (forces) in design. If you don't understand the forces at play (e.g., performance vs. flexibility), you may misuse the pattern and cause unintended consequences.

---

> 📝 **Tip:** Understand the *problem space* first. Then ask: "Is there a pattern that naturally solves this, while supporting the system’s constraints and quality goals?"

### Factory Method Pattern
The **Factory Pattern** is a **creational design pattern** that provides an interface for creating objects in a superclass but allows subclasses to alter the type of objects that will be created.

Instead of calling a constructor directly, you use a factory method to create an object.

#### 🧩 What Problem Does It Solve?

- You want to **encapsulate object creation logic**.
- You need to **create objects without specifying their exact class**.
- You want to **decouple the code that creates objects from the code that uses them**.

**Example Use Case:**
You have a payment system that supports multiple methods: `PayPal`, `CreditCard`, `Crypto`, etc. A factory can decide which payment object to create based on a parameter like `"paypal"` or `"credit"`.

#### ✅ When to Use

- When object creation is **complex** or involves **conditional logic**.
- When you have a **common interface or superclass**, but the actual object to instantiate can vary.
- When you want to **centralize and manage object creation** in one place.

#### 🚫 When *Not* to Use

- If you only need to instantiate a **single class** with no variation — the factory adds unnecessary complexity.
- If object creation is **simple** and doesn’t require customization.
- When you don't need polymorphism — creating objects directly is clearer and more readable.

---

#### Example 1: Simple Factory Pattern
This is simple and effective for small use case but not scalable.

##### Limitations
1. **Not scalable**: The get_pet() method hardcodes instances inside a dict. If you need to add new types or handle more complex creation logic (like configs, parameters), this approach gets messy.
2. **No common interface**: There’s no abstraction like Animal or Pet. So calling .speak() assumes all pets have that method — you’ll get runtime errors if not.
3. **No decoupling**: The caller still needs to know what strings to use ("dog", "cat"), and there’s no validation or fallback.

```python
class Dog:
    """ a simple dog class"""
    def __init__(self, name):
        self._name = name
    def speak(self):
        return "Woof!"
    
class Cat:
    """ a simple cat class"""
    def __init__(self, name):
        self._name = name
    def speak(self):
        return "Meow!"

# encapsulates object creation logic
def get_pet(pet="dog"):
    pets = dict(dog = Dog("Hope"), cat = Cat("Peace"))
    return pets[pet]

my_pet = get_pet("cat")
print(my_pet.speak())

my_pet = get_pet("dog")
print(my_pet.speak())
```

#### Example 2: Classic Factory Pattern

```python
from abc import ABC, abstractmethod

class Pet(ABC):
    # enforce PolyMorphism: derived class must implement this method
    @abstractmethod
    def speak(self): # this defines the interface that all pets must have
        print('pet sound')

# the below classes define behaviour
class Dog(Pet):
    def speak(self):
        return "Woof!"
class Cat(Pet):
    def speak(self):
        return "Meow!"

# this defines creation logic
class PetFactory():
    @staticmethod
    def get_pet(pet_type: str) -> Pet:
        pets = dict(dog=Dog(), cat=Cat())
        return pets[pet_type]
    # we can add more dependencies,config or logging here

my_pet = PetFactory.get_pet("dog")
print(my_pet.speak())

my_pet = PetFactory.get_pet("cat")
print(my_pet.speak())
```

### 🏭 Abstract Factor Pattern

### 🔹 What is it?

The **Abstract Factory Pattern** is a **creational design pattern** that provides an interface for creating **families of related or dependent objects** without specifying their concrete classes.

Unlike the Factory Method Pattern (which creates one object), **Abstract Factory creates multiple related objects** — all consistent with a particular theme, variant, or platform.

### 🧩 What Problem Does It Solve?

- You need to **create groups of related objects** (e.g., UI components, pets & accessories, OS-specific widgets).
- You want to **ensure compatibility** between created objects.
- You want to **switch entire object families easily** (e.g., from Windows UI to macOS UI).

### ✅ When to Use

- When your system needs to work with **multiple families of related products**.
- When you want to **enforce consistency** among related objects.
- When the product creation logic needs to be **decoupled and scalable**.

**Example Use Case:**
A GUI app that needs to support different operating systems. You can switch between `WindowsFactory` and `MacOSFactory` to get the correct `Button`, `Checkbox`, etc.

### 🚫 When *Not* to Use

- When you only need to create **a single type of object** (use Factory Method instead).
- When object families aren’t interdependent or related.
- When the number of product variations is small and unlikely to grow.

### 🧠 Core Structure

- **Abstract Factory**: Interface that declares creation methods for each type of product.
- **Concrete Factory**: Implements creation methods for a specific family of products.
- **Abstract Product**: Common interface for each kind of product.
- **Concrete Product**: Specific implementations of each product.
- **Client**: Uses the abstract factory and works only with interfaces.

> 📝 **Tip:** Think of Abstract Factory as **"Factory of Factories"**. It gives you a set of objects that are designed to work well together, with just one switch.

```python
from abc import ABC, abstractmethod

# define abstract product interface
class Pet(ABC):
    @abstractmethod
    def speak(self):
        pass
class PetToy(ABC):
    @abstractmethod
    def play(self):
        pass

# concrete products
class Dog(Pet):
    def speak(self):
        return "Woof!"
class DogToy(PetToy):
    def play(self):
        return "Dog plays with toy"
class Cat(Pet):
    def speak(self):
        return "Meow!"
class CatToy(PetToy):
    def play(self):
        return "Cat plays with toy"
    
# define abstract factory
class PetFactory(ABC):
    @abstractmethod
    def create_pet(self) -> Pet:
        pass
    @abstractmethod
    def create_pet_toy(self) -> PetToy:
        pass

# concrete factories
class DogFactory(PetFactory):
    def create_pet(self)  -> Pet:
        return Dog()
    def create_pet_toy(self) -> PetToy:
        return DogToy()

class CatFactory(PetFactory):
    def create_pet(self) -> Pet:
        return Cat()
    def create_pet_toy(self) -> PetToy:
        # return super().create_pet_toy()
        return CatToy()

## client code
def get_pet_environment(factory: PetFactory):
    pet = factory.create_pet()
    pet_toy = factory.create_pet_toy()
    print(f"Pet: {pet.speak()}")
    print(f"Pet Toy: {pet_toy.play()}")

get_pet_environment(DogFactory())
get_pet_environment(CatFactory())
```

### 🔒 Singleton Design Pattern

### 🔹 What is it?

The **Singleton Pattern** ensures that a class has **only one instance** and provides a **global point of access** to that instance.

It is a **creational design pattern**, often used when exactly one object is needed to coordinate actions across the system.


### 🧩 What Problem Does It Solve?

- Avoids creating multiple instances of a class that should only exist once (e.g., config manager, cpu instance, database connection, logging system).
- Centralizes access to shared resources.

### ✅ When to Use

- When only **one instance** of a class should exist across the application.
- When that instance manages shared resources or state (e.g., caching, settings, logging).
- When instantiating multiple copies would be **expensive or cause conflicts**.

### 🚫 When *Not* to Use

- When you need **multiple instances** for flexibility or testing.
- When using singletons just to access global variables (anti-pattern).
- When it introduces **hidden state or tight coupling** between components.

### 🧠 Common Real-World Examples

- Database connector
- Logger
- App config or environment manager
- Thread pool manager
---

```python
# example 1: singleton with only one instance

import threading # only required when we are using threading and want to lock for a thread safe singleton

class Singleton:
    _instance = None # this is class-level variable shared by all instances
    _lock = threading.Lock() # lock to make it thread safe

    # this method is called before __init__ method to allocate memory for a new instance
    # before initialization of the object, we make sure that 
    # if an instance is already created, we return that instance instead of creating a new one
    def __new__(cls):
        print(f"__new__ is called with cls = {cls}")
        with cls._lock:
            if cls._instance is None:
                print("Creating new instance...")
                cls._instance = super().__new__(cls)
        return cls._instance

    # this is not required but just to show that __init__ is called after __new__ by default
    def __init__(self):
        print(f"__init__ is called with self = {self}")

s1 = Singleton()
s2 = Singleton()
print(s1 is s2)
```

```python
# example 2: singleton with two instance limitation (a multi-ton)

class Multiton:
    _instances = [None, None]
    _max_instances = 2
    _next_index = 0

    def __new__(cls):
        if cls._next_index >= cls._max_instances:
            cls._next_index = 0
        instance = cls._instances[cls._next_index]
        if instance is None:
            instance = super().__new__(cls)
            cls._instances[cls._next_index] = instance
        cls._next_index += 1
        return instance
    
    def __init__(self):
        print(f"__init__ is called with self = {self}")
    
m1 = Multiton() # 1st instance
m2 = Multiton() # 2nd instance
m3 = Multiton() # repeat 1st instance
m4 = Multiton() # repeat 2nd instance
m5 = Multiton() # repeat 1st instance
m6 = Multiton() # repeat 2nd instance
print(m1 is m2)
print(m1 is m3)
print(m2 is m4)
print(m2 is m6)
print(m3 is m5)
```

Using metaclass, we can separate the logic for creation of singleton classes.

```python
# example 3: logger using singleton approach

### first create a metaclass to create singleton classes
class SingletonMeta(type): # type is the metaclass for all classes in Python
    _instances = {}
    def __call__(cls, *args, **kwargs):
        # if first time: create and store instance
        if cls not in cls._instances:
            # This calls the actual class (Logger's) constructors (__new__ and __init__)
            #   __new__ >>> allocate memory    and __init__  >>> initialize attributes
            instance = super().__call__(*args, **kwargs)
            cls._instances[cls] = instance  # store created instance
        return cls._instances[cls]


# create singleton logger class using singleton metaclass
# we use singleton metaclass to make sure only one instance of Logger is created
class Logger(metaclass=SingletonMeta):
    def __init__(self):
        self.log_file = "app.log"
        with open(self.log_file, "a") as f:
            f.write("Logger initialized\n")
    
    # interface for logging
    def log(self, message):
        with open(self.log_file, "a") as f:
            f.write(f"{message}\n")

# client code
logger1 = Logger() # when logger1 is created, singleton metaclass creates Logger instance which is stored in _instances
logger1.log("This is logger1")
logger2 = Logger() # when logger2 is created, singleton metaclass returns the stored Logger instance
logger2.log("This is logger2")
print(logger1 is logger2)
```

```python
# using the above singleton metaclass, we will try creating another singleton class called db connection
class DBConnection(metaclass=SingletonMeta):
    def __init__(self):
        self.db_file = "app.db"
        with open(self.db_file, "a") as f:
            f.write("DB Connection initialized\n")
    
    def execute(self, query):
        with open(self.db_file, "a") as f:
            f.write(f"{query}\n")

# client code
db1 = DBConnection()
db1.execute("SELECT * FROM users")
db2 = DBConnection()
db2.execute("SELECT * FROM orders")
print(db1 is db2)
```

```python
# BORG: a variant of singleton where all instances share the same state
# this is useful when we want to share the state across multiple instances

class Borg:
    _shared_state = {} # this is class-level variable shared by all instances of this class
    def __init__(self):
        # __dict__ is a build-in dictionary that holds all the attributes of an object
        self.__dict__ = self._shared_state # this makes all instances share the same state

class Config(Borg):
    def __init__(self, **kwargs):
        super().__init__() # call the parent class's __init__ method
        self.__dict__.update(kwargs) #apply new attributes to the shared state

c1 = Config() # default setting
c2 = Config(setting="dark mode", user="admin") # dark mode setting

print(c1.setting) # although c1 is on default setting, it will show dark mode because c2 is on dark mode
print(c2.setting) # dark mode
print(c1 is c2)

# the attributes are shared
print(c1.__dict__)
print(c2.__dict__)

print(c1.setting is c2.setting)
```
